```python
# Writing a simple text file
filename1 = "example1.txt"
filename2 = "example2.txt"
content1 = "Hello there"
content2 = "Salut ici"

with open(filename1, "w") as fobj:
    fobj.write(content1)
with open(filename2, "w") as fobj:
    fobj.write(content2)

# Rename is forced even if the target exists
os.rename(filename1, filename2)

with open(filename2, "r") as fobj:
    content = fobj.read()

assert content == content1

print("Ok")
```
