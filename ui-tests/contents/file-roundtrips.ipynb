```python
# Writing a simple text file
filename = "example.txt"
content = "Hello there"

with open(filename, "w") as fobj:
    fobj.write(content)

with open(filename, "r") as fobj:
    assert fobj.read() == content

os.unlink(filename)

print("Ok")
```

```python
# Writing a binary image, and reading it
from PIL import Image
import numpy as np

filename = "large_image.png"

# Target size: adjust if needed
width = 5000
height = 500

# Create random pixel data (prevents compression)
data = np.random.randint(0, 256, (height, width, 3), dtype=np.uint8)

# Create image
img = Image.fromarray(data).convert("RGB")

# Save as PNG (low compression = bigger file)
img.save(filename, format="PNG", compress_level=0)

from PIL import Image
import numpy as np

# Load the saved image
loaded_img = Image.open(filename)

# Convert back to numpy array
loaded_data = np.array(loaded_img)

# Assert same shape
assert loaded_data.shape == data.shape, "Shape mismatch!"

# Assert pixel-by-pixel equality
assert np.array_equal(loaded_data, data), "Image data does not match!"

os.unlink(filename)

print("Ok")
```

```python
# Writing a binary file
from pathlib import Path

name = "noext-binary"

original = bytes([0xFF, 0xFE, 0xFD, 0x80, 0x61])

p = Path(name)
p.write_bytes(original)
roundtrip = Path(name).read_bytes()

assert roundtrip == original

os.unlink(name)

print("Ok")
```

```python
# Writing a binary file with no extension
from pathlib import Path

name = "ext-binary"

original = bytes([0xFF, 0xFE, 0xFD, 0x80, 0x61])

p = Path(name)
p.write_bytes(original)
roundtrip = Path(name).read_bytes()

assert roundtrip == original

os.unlink(name)

print("Ok")
```
