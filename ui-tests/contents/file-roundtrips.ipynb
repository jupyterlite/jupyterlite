```python
# All the following cells should print Ok.
import os

print("Ok")
```

```python
# Writing a simple text file
filename = "example.txt"
content = "Hello there"

with open(filename, "w") as fobj:
    fobj.write(content)

with open(filename, "r") as fobj:
    assert fobj.read() == content

os.unlink(filename)

print("Ok")
```

```python
# Writing a simple text file with no extension
filename = "example"
content = "Coucou toi"

with open(filename, "w") as fobj:
    fobj.write(content)

with open(filename, "r") as fobj:
    assert fobj.read() == content

os.unlink(filename)

print("Ok")
```

```python
# Writing a binary file
from pathlib import Path

name = "binary.bin"
original = bytes([0xFF, 0xFE, 0xFD, 0x80, 0x61])

Path(name).write_bytes(original)
roundtrip = Path(name).read_bytes()

assert roundtrip == original

os.unlink(name)

print("Ok")
```

```python
# Writing a binary file with no extension
from pathlib import Path

name = "noext-binary"
original = bytes([0xFF, 0xFE, 0xFD, 0x80, 0x61])

Path(name).write_bytes(original)
roundtrip = Path(name).read_bytes()

assert roundtrip == original

os.unlink(name)

print("Ok")
```
