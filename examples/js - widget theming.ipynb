# Theming JupyterLite with Widgets

```javascript
let {FloatSlider, HTML, dlink, Checkbox} = kernel.widgets
self.make_a_theme_control = (v, t, k=FloatSlider) => {
    s = k({description: v, ...(t || {})})
    h = HTML()
    dlink([s, 'value'], [h, 'value'], (s) => `<style>:root{ ${v}: ${s}em; }</style>`)
    h.display()
    s.display()
}
self.make_a_color_picker = (v, rgba="rgba") => {
    const S = rgba.split("").map((d) => FloatSlider({description: d, min: 0, max: d == 'a' ? 1 : 255, step: 0.01}))
    S.map((s) => s.display())
    S.map((s) => s.observe(()=> {
        const style = `${v}: ${rgba}(${S.map(({value})=> value).join(',')});`;
        h.value = `<style>:root{ ${style} }</style><pre>${style}</pre>`
    }))
    h = HTML()
    h.display()
}
```

```javascript
make_a_theme_control('--jp-cell-collapser-width');
```

```javascript
make_a_theme_control('--jp-notebook-padding');
```

```javascript
self.make_a_color_picker('--jp-cell-editor-background')
```
